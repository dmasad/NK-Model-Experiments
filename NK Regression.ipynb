### Convert space and fitness to DataFrame

```python
model_space = []
for key, val in model.enumerate_space().items():
    row = {}
    for i in range(model.n):
        row["x" + str(i)] = key[i]
    row["Fitness"] = val
    model_space.append(row)

df = pandas.DataFrame(model_space)
```

```python
df.head()
```

## Fit linear model

```python
df["Constant"] = 1
```

```python
x_cols = df.columns[1:]
```

```python
lm = sm.OLS(df.Fitness, df[x_cols])
fit = lm.fit()
```
